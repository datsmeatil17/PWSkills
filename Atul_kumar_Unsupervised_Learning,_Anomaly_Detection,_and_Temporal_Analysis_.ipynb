```python
'''


##Question 1: What is Dimensionality Reduction? Why is it important in machine learning?
Dimensionality reduction is the process of transforming a dataset with many features (dimensions) into one with fewer dimensions while retaining as much of the relevant information as possible. In effect, you go from a high-dimensional space to a lower-dimensional representation.

 It is important in machine learning for several reasons: firstly, high-dimensional data often suffer from the so-called “curse of dimensionality” — as the number of features grows, data become sparse, distances lose meaning, and generalisation becomes harder.

 Secondly, many algorithms become computationally expensive or overfit when there are many irrelevant or redundant features. Dimensionality reduction helps reduce computing cost and model complexity.
GeeksforGeeks
 Thirdly, reducing dimensions can improve visualization (e.g., projecting into 2D/3D), noise removal, and help uncover latent structure in the data.
GeeksforGeeks
 Thus, using dimensionality reduction as a preprocessing or embedding step helps make downstream ML models more efficient, interpretable, and robust.

##Question 2: Name and briefly describe three common dimensionality reduction techniques.
Here are three widely used techniques:

Principal Component Analysis (PCA): A linear technique that projects the data onto orthogonal directions (principal components) which capture maximum variance, thereby reducing dimensions while retaining as much variance as possible.


t-Distributed Stochastic Neighbor Embedding (t-SNE): A nonlinear technique useful mostly for visualization: it converts high dimensional distances to pairwise probabilistic similarities, then finds a low-dimensional embedding so that similar points stay close and dissimilar points are farther apart.


Independent Component Analysis (ICA): A linear method (but with a different aim than PCA) that separates multivariate signal into statistically independent non-Gaussian components. It’s useful when the underlying sources are assumed independent.

 (One might also mention Uniform Manifold Approximation & Projection (UMAP) or Autoencoders, but three were requested.)

##Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.
Clustering is a technique in unsupervised learning that groups unlabeled data points into clusters such that points within the same cluster are more similar (by some metric) to each other than to points in other clusters.

 Because there are no labels, the goal is to uncover natural groupings or hidden structure in the data. Three popular clustering algorithms:

K-Means: A partitioning (centroid-based) method where you pre-choose the number of clusters K, assign points to the nearest centroid, then update centroids iteratively until convergence.


Hierarchical Clustering: Can be agglomerative (start with each point as its own cluster and merge step by step) or divisive; produces a tree (dendrogram) of clusters.


DBSCAN (Density-Based Spatial Clustering of Applications with Noise): A density-based method which groups points densely connected and marks low-density regions (sparse points) as noise/outliers. It handles irregular shapes well and doesn’t require specifying K.


##Question 4: Explain the concept of anomaly detection and its significance.
Anomaly detection (also known as outlier detection) is the process of identifying observations or events that deviate significantly from the majority of the data — anomalies may indicate rare but important occurrences, such as fraud, fault, intrusion, or other unusual behaviour.

 The significance lies in its many applications: in cybersecurity (detecting intrusions), in finance (fraudulent transactions), in manufacturing/IoT (equipment faults), in health (disease outbreaks) and more. By detecting anomalies early, one can respond timely to risk, uncover hidden opportunities (unusual patterns), or clean data (removing outliers before modelling). Because anomalies are by definition rare/unlabelled, anomaly-detection methods often must work without much labelled data or supervision — making them a critical unsupervised or semi-supervised tool in ML.

##Question 5: List and briefly describe three types of anomaly detection techniques.
Here are three common techniques:

Statistical / Modeling-based methods: These build a model of “normal” behaviour (e.g., using Gaussian distributions, thresholding) and flag data points that deviate significantly. For example, setting thresholds on residuals after trend-seasonality decomposition in time series.
GeeksforGeeks

Tree- or ensemble-based methods (e.g., Isolation Forest): For example, Isolation Forest works by recursively partitioning data and measuring how quickly a point becomes isolated — anomalies tend to be isolated with fewer splits.


Proximity / density-based methods (e.g., Local Outlier Factor, LOF): These compute a point’s local density relative to its neighbours; points with much lower density than neighbours are flagged as anomalies.

 (Additionally, one can mention unsupervised deep-learning methods like autoencoders trained on normal data and using reconstruction error for anomaly scoring.)

##Question 6: What is time series analysis? Mention two key components of time series data.
Time series analysis refers to techniques for analysing data points collected sequentially over time (often at regular intervals) to understand underlying patterns (trend, seasonality, cycles, etc.), forecast future values, detect changes or anomalies, and more. A time series is thus more than just independent samples — temporal order, autocorrelation, and temporal patterns matter. Two key components of time series data are:

Trend: The long-term movement or direction in the data (e.g., upward growth or downward decline) over time.

Seasonality (or seasonal component): Regular, predictable fluctuations that occur at fixed intervals (e.g., daily, weekly, yearly) often driven by repeating, known factors.


##Question 7: Describe the difference between seasonality and cyclic behaviour in time series.
Seasonality refers to patterns that repeat at fixed, known intervals (for example, monthly sales peaks in December, daily traffic peaks in the evening, etc.). The period is stable and the patterns are predictable.

 Cyclic behaviour, by contrast, refers to fluctuations that occur over irregular or longer time spans (e.g., economic business cycles spanning several years) and the timing or amplitude of the cycles may not be fixed or predictable.

 So the key difference: seasonality = fixed periodicity and predictable pattern; cyclic = variable length, less predictable periodicity, often driven by broader external factors rather than fixed calendar events.'''






```

```python
#Question 8: Write Python code to perform K-means clustering on a sample dataset. (Include your Python code and output in the code box below.)

# Sample Python code using scikit-learn to perform K-Means clustering
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# generate sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# fit KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
y_pred = kmeans.fit_predict(X)

# plot results
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap='viridis')
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=100, alpha=0.9, marker='X')
plt.title('K-Means clustering result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


'''Sample Output:
A scatter plot showing 300 points colored by their assigned cluster (4 clusters), and red “X” markers showing the cluster centroids. Points group around the centroids and clusters are clearly visibly separated.'''
```

```python
#Question 9: What is inheritance in OOP? Provide a simple example in Python.
#Inheritance in object-oriented programming (OOP) is the mechanism by which one class (the subclass or derived class) inherits attributes and methods from another class (the base class or parent class). This promotes code reuse and hierarchical relationships. A subclass can extend or override behaviour of the base class.
#Example:

# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

# Subclass inheriting from Animal
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    # override speak method
    def speak(self):
        return f"{self.name} the {self.breed} says Woof!"

# usage
d = Dog("Buddy", "Golden Retriever")
print(d.speak())
# output: Buddy the Golden Retriever says Woof!


#Here, Dog inherits from Animal (so it reuses __init__ for name via super(), and inherits other base-class methods). It then overrides speak() to provide specific behaviour.
```
