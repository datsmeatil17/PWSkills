```python
'''

Question 1: What is hypothesis testing in statistics?
In hypothesis testing, two competing hypotheses are formulated:

Null Hypothesis: Poses no effect or no difference, serving as the default assumption.

Alternative Hypothesis : Contradicts the null hypothesis, suggesting the presence of an effect or difference.

The goal is to evaluate the sample data to determine whether there is enough statistical evidence to reject the null hypothesis in favor of the alternative hypothesis. This process involves calculating a test statistic and comparing it to a critical value or using a p-value to assess the strength of the evidence .

Question 2: What is the null hypothesis, and how does it differ from the alternative
hypothesis?
In statistical hypothesis testing, the null hypothesis (H₀) and the alternative hypothesis (H₁ or Ha) are two mutually exclusive statements that provide a framework for evaluating evidence in data.

Null Hypothesis (H₀)
The null hypothesis posits that there is no effect, no difference, or no relationship between variables in the population. It serves as the default assumption that any observed differences are due to random chance. Researchers aim to test whether there is sufficient evidence to reject this assumption.

Alternative Hypothesis (H₁ or Ha)
The alternative hypothesis suggests that there is a statistically significant effect, difference, or relationship between variables. It is what researchers seek to support through evidence, indicating that the null hypothesis may not adequately explain the observed data.

Question 3: Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
In hypothesis testing, the significance level (denoted as α) is a critical threshold that helps researchers determine whether the observed data provide sufficient evidence to reject the null hypothesis. It quantifies the probability of making a Type I error—incorrectly rejecting a true null hypothesis. Commonly set at 0.05, this means there's a 5% risk of concluding that an effect exists when it doesn't.
The significance level serves as a benchmark for evaluating the p-value obtained from the test statistic. If the p-value is less than or equal to α, the null hypothesis is rejected, indicating that the observed effect is statistically significant. Conversely, if the p-value exceeds α, the null hypothesis is not rejected, suggesting insufficient evidence to support the alternative hypothesis.

Question 4: What are Type I and Type II errors? Give examples of each.
In statistical hypothesis testing, Type I and Type II errors represent two potential mistakes made when making decisions about the null hypothesis. Understanding these errors is crucial for interpreting test results and making informed decisions.
Type I Error (False Positive)
Definition: Occurs when the null hypothesis is true, but is incorrectly rejected.

Consequence: A false positive—believing there is an effect or difference when there isn't.

Example: In a criminal trial, convicting an innocent person.

Probability: Denoted by α (alpha), commonly set at 0.05.
Type II Error (False Negative)
Definition: Occurs when the null hypothesis is false, but is incorrectly not rejected.

Consequence: A false negative—failing to detect an effect or difference when one exists.

Example: In medical testing, failing to detect a disease that is present.
Probability: Denoted by β (beta), and the power of a test is 1 − β.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use
each.
In statistical hypothesis testing, both the Z-test and T-test are used to determine if there is a significant difference between sample means and population means or between two sample means. The choice between them depends primarily on the sample size and whether the population standard deviation is known.

Z-Test: Use when the sample size is large (typically n ≥ 30) and the population standard deviation is known. This scenario allows for the application of the standard normal distribution to assess the significance of the sample mean.

T-Test: Use when the sample size is small (typically n < 30) or the population standard deviation is unknown. The t-distribution accounts for the additional uncertainty in estimating the population standard deviation from the sample.



'''

```

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and
#p=0.5, then plot its histogram.

#(Include your Python code and output in the code box below.)
#Hint: Generate random number using random function.

import random
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate binomial distribution
binomial_data = [sum(random.random() < p for _ in range(n)) for _ in range(1000)]

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.grid(True)
plt.show()
```

```python
'''Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''



import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 50.3, 50.1, 50.2, 50.4,
               50.0, 50.6, 50.2, 50.1, 50.3, 50.5, 50.0, 50.7, 50.3, 50.1,
               50.2, 50.4, 50.0, 50.6, 50.2, 50.1, 50.3, 50.5, 50.0, 50.7]

# Population parameters
population_mean = 50
population_std_dev = 2
n = len(sample_data)

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)  # Sample standard deviation

# Z-statistic calculation
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(n))

# P-value calculation for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Decision based on significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average daily sales are significantly different from 50.")
else:
    print("Fail to reject the null hypothesis: The average daily sales are not significantly different from 50.")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
mean = 50
std_dev = 5
sample_size = 1000
confidence_level = 0.95

# Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)  # Sample standard deviation

# Calculate the standard error of the mean
sem = sample_std_dev / np.sqrt(sample_size)

# Calculate the 95% confidence interval
ci_lower, ci_upper = stats.norm.interval(confidence_level, loc=sample_mean, scale=sem)

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f'CI Lower: {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f'CI Upper: {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean. '''

import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 50.3, 50.1, 50.2, 50.4,
        50.0, 50.6, 50.2, 50.1, 50.3, 50.5, 50.0, 50.7, 50.3, 50.1,
        50.2, 50.4, 50.0, 50.6, 50.2, 50.1, 50.3, 50.5, 50.0, 50.7]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-scores
z_scores = [(x - mean) / std_dev for x in data]

# Plot histogram of Z-scores
plt.figure(figsize=(8, 6))
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Z-Scores')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
