```python
#Question 9: Question 9: Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
size = 1000
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Compute the mean and standard deviation
computed_mean = np.mean(data)
computed_std_dev = np.std(data)

# Display the computed mean and standard deviation
print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std_dev:.2f}")

# Plot the histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Question 10: You are working as a data analyst for a retail company. The company has
#collected daily sales data for 2 years and wants you to identify the overall sales trend.
#daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
#235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
#● Explain how you would apply the Central Limit Theorem to estimate the average sales
#with a 95% confidence interval.
#● Write the Python code to compute the mean sales and its confidence interval.

#To analyze the daily sales data and estimate the average sales with a 95% confidence interval, we can apply the Central Limit Theorem (CLT). The CLT states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population's distribution shape
#Steps to Apply the CLT:
#Calculate the Sample Mean and Standard Deviation: Compute the mean and standard deviation of the daily sales data.

#Determine the Sample Size: The number of data points in the sample, which is 20 in this case.

#Compute the Standard Error: Divide the sample standard deviation by the square root of the sample size.

#Find the Z-Score for 95% Confidence: For a 95% confidence level, the Z-score is approximately 1.96.

#Calculate the Margin of Error: Multiply the Z-score by the standard error.

#Determine the Confidence Interval: Subtract and add the margin of error from the sample mean to get the lower and upper bounds of the confidence interval.

import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Calculate sample mean and standard deviation
avg_sales = np.mean(daily_sales)
std_dev_sales = np.std(daily_sales, ddof=1)  # Sample standard deviation
n = len(daily_sales)

# Calculate standard error
standard_error = std_dev_sales / np.sqrt(n)

# Z-score for 95% confidence
z_score = stats.norm.ppf(0.975)

# Margin of error
margin_of_error = z_score * standard_error

# Confidence interval
lower_bound = avg_sales - margin_of_error
upper_bound = avg_sales + margin_of_error

print(f"Sample Mean: {avg_sales:.2f}")
print(f"Sample Standard Deviation: {std_dev_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
