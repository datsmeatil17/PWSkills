```python
'''Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Descriptive statistics summarize actual data using metrics like mean, median, mode, range, and standard deviation. ex- If Class A students scored 70, 85, 90, and 65, descriptive statistics help compute the average (77.5) and range (25).

Inferential Statistics- Uses sample data to draw conclusions or make predictions about a broader population, acknowledging uncertainty through probability theory. ex-  A sample of 100 students might be used to estimate the average grade of all students in a country.

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Sampling in statistics is the process of selecting a subset (sample) from a larger population to estimate population characteristic saving time and cost when surveying everyone is not feasible.

Random sampling (or simple random sampling) gives every individual an equal, independent chance of being selected, reducing bias and enabling generalization
Stratified sampling involves dividing the population into non‑overlapping subgroups (strata) based on shared traits, then randomly sampling each subgroup—ensuring all segments are accurately represented and yielding more precise estimates

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Definitions
Mean (arithmetic mean): the sum of all values divided by the number of observations.
Median: the middle value when data are ordered. For odd-sized samples, it's the center; for even-sized, it's the average of the two middle values.
Mode: the value that occurs most frequently. It applies to numerical, ordinal, and also nominal (categorical) data

Importance-
Mean uses all data points, making it powerful for symmetric distributions.
Median resists distortion by outliers, offering a truer center for skewed data
Mode uniquely handles categorical or nominal data—for instance, identifying the most common surname or preference

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?
Skewness measures the asymmetry of a data distribution around its mean — whether the values are more stretched out on one side. A positive skew (or right-skew) means the right tail is longer, with a few higher-end outliers pulling the mean above the median and mode  This implies the data is concentrated toward the lower values with some unusually large values. For example, income distributions often exhibit positive skew: most people earn moderate amounts, but a few earn significantly more
Kurtosis
Kurtosis indicates the "tailedness" of a distribution—the likelihood of extreme values or outliers—rather than the height of its peak  A normal (mesokurtic) distribution has kurtosis near zero (with excess kurtosis standardized to zero)
Leptokurtic distributions (positive kurtosis) have heavy tails—more extreme outliers—while platykurtic distributions (negative kurtosis) have lighter tails, with fewer outliers '''
```

```python
# Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] (Include your Python code and output in the code box below.


import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Using numpy
cov_matrix = np.cov(list_x, list_y, ddof=1)  # sample covariance
cov_xy = cov_matrix[0, 1]
corr_xy = np.corrcoef(list_x, list_y)[0, 1]

print("Using NumPy:")
print("Covariance (sample):", cov_xy)
print("Pearson correlation:", corr_xy)

# Manual calculation
mean_x = sum(list_x) / len(list_x)
mean_y = sum(list_y) / len(list_y)
cov_manual = sum((x - mean_x)*(y - mean_y) for x, y in zip(list_x, list_y)) / (len(list_x) - 1)
std_x = (sum((x - mean_x)**2 for x in list_x) / (len(list_x) - 1)) ** 0.5
std_y = (sum((y - mean_y)**2 for y in list_y) / (len(list_y) - 1)) ** 0.5
corr_manual = cov_manual / (std_x * std_y)

print("\nManual calculation:")
print("Covariance (sample):", cov_manual)
print("Pearson correlation:", corr_manual)
```

```python
# Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] (Include your Python code and output in the code box below.)

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Dataset
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a boxplot
sns.boxplot(data=data)
plt.title("Boxplot of Given Data")
plt.show()

# Calculate IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Determine outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower bound for outliers: {lower_bound}")
print(f"Upper bound for outliers: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
# Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000] (Include your Python code and output in the code box below.)

'''Covariance measures the directional relationship between two paired samples. It indicates whether increases in one variable correspond to increases or decreases in another. A positive covariance suggests that as advertising spend increases, daily sales tend to increase as well. However, covariance is sensitive to the scale of the variables, making it difficult to interpret without context.
Correlation
Correlation, specifically the Pearson correlation coefficient, normalizes covariance by the standard deviations of the variables, resulting in a dimensionless value between -1 and +1. A value close to +1 indicates a strong positive linear relationship, meaning that as advertising spend increases, daily sales also increase proportionally. This makes correlation more interpretable and comparable across different datasets.'''

import numpy as np
import matplotlib.pyplot as plt

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute Pearson correlation coefficient
correlation_matrix = np.corrcoef(advertising_spend, daily_sales)
correlation_coefficient = correlation_matrix[0, 1]

# Output the result
print(f"Pearson Correlation Coefficient: {correlation_coefficient:.4f}")

# Scatter plot for visualization
plt.scatter(advertising_spend, daily_sales)
plt.title("Advertising Spend vs Daily Sales")
plt.xlabel("Advertising Spend ($)")
plt.ylabel("Daily Sales ($)")
plt.grid(True)
plt.show()
```

```python
# Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use. ● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7] (Include your Python code and output in the code box below.)

'''Summary Statistics
1. Mean: The average score, providing a central value.
2. Median: The middle value when data is ordered, indicating the central tendency.
3. Mode: The most frequently occurring score, highlighting common responses.
4. Standard Deviation: Measures the spread of scores around the mean, indicating variability.
5. Range: The difference between the highest and lowest scores, showing the spread of data.
Visualization: Histogram
A histogram provides a graphical representation of the frequency distribution of the data, helping to identify patterns such as skewness, modality, and the presence of outliers. '''

import numpy as np
import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
mode_score = max(set(survey_scores), key=survey_scores.count)
std_dev = np.std(survey_scores)
data_range = np.ptp(survey_scores)

# Display summary statistics
print(f"Mean: {mean_score}")
print(f"Median: {median_score}")
print(f"Mode: {mode_score}")
print(f"Standard Deviation: {std_dev}")
print(f"Range: {data_range}")

# Create histogram
plt.hist(survey_scores, bins=range(1, 12), edgecolor='black', alpha=0.7)
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```
