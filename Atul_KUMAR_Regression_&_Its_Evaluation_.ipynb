```python
'''

#Question 1: What is Simple Linear Regression?

Simple Linear Regression is a foundational statistical technique used to explore and model the relationship between two numerical variables—one considered the predictor and the other the outcome. At its core, it seeks to represent how, on average, the outcome variable tends to change when the predictor variable changes. Importantly, this relationship is captured using the simplest possible form—that is, a straight-line trend—not more complex or curved patterns.

The primary purpose of this method is twofold: to understand how strongly the predictor influences the outcome, and to predict what the outcome might be for a given value of the predictor. For example, you could investigate how study time affects test scores or how advertising spend relates to sales.


To find that best‑fit straight-line trend, the method typically uses the "least squares" approach—choosing the line that minimizes the differences between observed outcomes and the line’s predictions.

For this model to be reliable, certain conditions (assumptions) should hold:

The relationship should genuinely follow a straight‑line trend.

Variability in prediction errors should remain consistent across values.

Observations must be independent of each other.

Often, a roughly normal distribution of residual differences is also assumed.

Because of its simplicity, this regression is highly interpretable and widely used across fields like economics, biology, and business. It provides a clear, easy-to-understand "first look" at how one variable affects another.


#Question 2: What are the key assumptions of Simple Linear Regression?

Simple Linear Regression rests on four fundamental assumptions to ensure its results are reliable and interpretable:

Linearity
This means the connection between the predictor and the outcome follows a straight-line pattern. In other words, as one variable moves, the other tends to move in a directly proportional way—not in curves or irregular jumps. This assumption is foundational—if the relationship is not straight-line, the model’s conclusions can be misleading.

Independence of Observations (or Errors)
Each data point should be independent—one observation shouldn’t influence another. Likewise, the errors (the differences between actual outcomes and model predictions) should not have hidden patterns or relationships. If observations or errors are related—such as in time‑series data—then specialized techniques may be needed instead.


Homoscedasticity (Constant Variance)
The spread of prediction errors should be consistent across all predictor values. That is, the variability around the model's predicted line should be roughly the same—whether the predictor is low or high. If errors get systematically larger or smaller in some ranges, the model’s confidence in its predictions may be skewed.


Normality of Errors
The deviations of actual outcomes from the predicted ones (the residuals) ideally should follow a bell‑shaped, "normal" distribution. This is especially important when making statistical inferences—like confidence intervals or p‑values—but is less critical for prediction alone.


Practical Angle – When Things Go Off Track

If these assumptions don’t hold—say, if the relationship is curved or the error spread widens—you might still get biased or inefficient estimates, or invalid statistical conclusions. Practitioners often check these assumptions visually using diagnostic tools like plots of residuals or Q‑Q plots.

As discussed in communities like r/AskStatistics, these assumptions are what make Ordinary Least Squares estimators unbiased and efficient, and they enable valid hypothesis testing and inference.


In summary, Simple Linear Regression works best when:

the relationship is straight,

observations stand alone,

errors are consistent in spread,

and errors are shaped normally.

#Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

Heteroscedasticity refers to a situation in regression where the spread—or scatter—of prediction errors (residuals) varies across different levels of the independent variable. Instead of having a consistent clustering of errors, you might see them spreading out or narrowing as the predictor values change
. Imagine modeling income vs. spending: low-income groups might have tightly grouped spending patterns, while high-income groups exhibit much wider variability—this is a typical example of heteroscedasticity

Why It Matters

Unbiased but inefficient estimates: Regression methods like OLS still yield unbiased estimates of relationships, but they no longer guarantee the most precise (lowest variance) results. In other words, the estimates aren’t as reliable as they could be

.

Inaccurate standard errors and inference: Since residual variability isn’t uniform, the standard errors of coefficient estimates become unreliable. This compromises hypothesis tests, confidence intervals, and p-values—they can be either overstated or understated—leading to faulty conclusions about which predictors are significant

Loss of estimator efficiency (Gauss‑Markov breakdown): With heteroscedasticity, the Gauss‑Markov theorem no longer holds, meaning OLS is not the Best Linear Unbiased Estimator anymore


From forums like r/econometrics:

“Robust standard errors will (generally) be larger than the non‑robust errors... Using them is therefore simply more conservative and gives you more certainty”


And from r/AskStatistics:

“Variance is inflated... If variance is too high… you cannot reject the null hypothesis... This is a Type II error.”

These insights underscore that ignoring heteroscedasticity can underestimate uncertainty and cause you to miss meaningful effects.

Summary

Definition: Heteroscedasticity occurs when error variability changes across observations.

Importance:

Estimates remain unbiased—but are less precise.

Standard errors and statistical tests may be misleading.

OLS loses its optimal efficiency property.

Impact:

Risk of incorrect inferences, like false positives or negatives.

Inflated uncertainty leads to weaker testing power.


#Question 4: What is Multiple Linear Regression?

Multiple Linear Regression (MLR) is a statistical approach used to understand or predict a continuous outcome based on two or more independent variables. Unlike simple linear regression, which involves just one predictor, MLR accounts for multiple influences simultaneously, offering a more realistic view of complex systems


At its heart, MLR does two things:

Explains relationships: It shows how each predictor relates to the outcome, while controlling for the others—revealing the individual influence of each variable


Enhances prediction: By incorporating multiple variables, MLR generally provides more accurate predictions compared to relying on a single factor


This method is widely applied in fields like economics, finance, marketing, biology, and social sciences, where outcomes are shaped by several factors at once


However, MLR has important caveats:

It assumes each predictor relates linearly to the outcome and that these relationships are separate—so that effects don’t get mixed up

Multicollinearity, or high correlation among predictors, can muddle the interpretation of individual variables, making it hard to pinpoint which factor matters most


Adding too many predictors can lead to overfitting, where the model performs well on the known data but poorly on new, unseen data


In practice, analysts use statistical software to fit MLR models and assess how well the predictors collectively explain variation in the outcome, often using a measure like R‑squared to evaluate model strength

In summary: Multiple Linear Regression is a powerful tool that captures how multiple factors jointly impact a continuous outcome, providing clearer insights and better predictive power—so long as its assumptions are respected and the model is carefully managed against pitfalls like multicollinearity and overfitting.


#Question 5: What is polynomial regression, and how does it differ from linear
regression?


Polynomial regression is a modeling technique used when the relationship between an input (predictor) and output (response) variable is curved or more complex than a straight line. Instead of fitting a simple trajectory, it adapts to the shape of the data by allowing the fitting line to bend and follow non‑linear patterns. Statisticians still treat it as a form of linear regression because the model estimates remain linear in the parameters (i.e., in the coefficients) even though the overall shape can be curved.


In contrast, linear regression assumes a consistent, unchanging relationship—where an increase in the predictor leads to a constant change in the outcome. It's straightforward to interpret and works well when data follow that simple, direct trend.


Here are the key differences:

Flexibility: Polynomial regression can model wavy or curved trends, capturing complex patterns in the data, while linear regression is limited to straight-line relationships.


Complexity and overfitting risk: Adding more curvature (higher-degree terms) makes the model flexible but also more complex and sensitive—leading to risks like overfitting, where the model starts chasing noise rather than genuine patterns.


Interpretability: With linear regression, you can easily explain how the predictor affects the response. Polynomial regression introduces multiple terms (like squared or cubic versions of the predictor), making interpretation less intuitive and more abstract.

'''
```

```python
#Question 6:  Implement a Python program to fit a Simple Linear Regression model to
#the following sample data:
#● X = [1, 2, 3, 4, 5]
#● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
#Plot the regression line over the data points.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Predictor values as a 2D array
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])       # Response values

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Generate predictions for plotting
Y_pred = model.predict(X)

# Visualize data points and fitted regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression Fit')
plt.legend()
plt.show()
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
#● Area = [1200, 1500, 1800, 2000]
#● Rooms = [2, 3, 3, 4]
#● Price = [250000, 300000, 320000, 370000]
#Check for multicollinearity using VIF and report the results.


import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Sample data
data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

# Prepare predictors (X) and target (y)
X = data[['Area', 'Rooms']]
y = data['Price']

# Add constant term for intercept
X_with_const = sm.add_constant(X)

# Fit Multiple Linear Regression model
model = sm.OLS(y, X_with_const).fit()
print("Regression Summary:\n", model.summary())

# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data['Feature'] = X_with_const.columns
vif_data['VIF'] = [
    variance_inflation_factor(X_with_const.values, i)
    for i in range(X_with_const.shape[1])
]
print("\nVariance Inflation Factors (VIF):\n", vif_data)
```

```python
#Question 8:  Implement polynomial regression on the following data:
#● X = [1, 2, 3, 4, 5]
#3
#● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
#Fit a 2nd-degree polynomial and plot the resulting curve.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Generate polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict results for plotting
Y_pred = model.predict(X_poly)

# Plot data points and the fitted polynomial curve
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Quadratic fit')
plt.title('Polynomial Regression (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
# X = [10, 20, 30, 40, 50]
#● Y = [15, 35, 40, 50, 65]
#Assess heteroscedasticity by examining the spread of residuals.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit the model
model = LinearRegression()
model.fit(X, Y)

# Compute predictions and residuals
y_pred = model.predict(X)
residuals = Y - y_pred

# Plot residuals vs. predicted values
plt.scatter(y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.show()
```
